```python
import pandas as pd
```

```python
df  = pd.read_csv('SampleData.csv',header=None)
```

```python
df.head()
```

## Dropping the label

```python
df_1 = df.drop(0,axis=1)
```

```python
df_1
```

## Number of features



```python
df_1.shape
```

No of columns is the features(As I've dropped the label).
Hence number of features is 750.


The order of data matrix 250 X 751 

## The function to calculate and similarity for a pair of rows a and b 

```python
def sim(a,b):
    match = 0
    for i in range(1,750):
        if df_1.loc[int(a),int(i)] == df_1.loc[int(b),int(i)]:
            match = match +1
    similarity = match/750
    return similarity
        
    
    
```

```python
similarity_matrix = []
```

```python
for i in range(250):
    temp = []
    for j in range(250):
        if i!=j:
            similarity = sim(i,j)
            temp.append(similarity)
        else:
            temp.append(0)
    similarity_matrix.append(temp)
```

## Writing the Similarity matrix into a csv file

```python
import csv
```

```python
with open('sMatrix.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for i in range(250):
        writer.writerow(similarity_matrix[i])
```
